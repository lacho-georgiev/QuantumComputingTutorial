```python
from qiskit import *
from math import pi
qr = QuantumRegister(4)
cr = ClassicalRegister(3)
circuit = QuantumCircuit(qr, cr)
```

```python
circuit.h(0)
circuit.h(1)
circuit.h(2)
circuit.x(3)
circuit.cp(pi/3, 0, 3)
circuit.cp(pi/3, 1, 3)
circuit.cp(pi/3, 1, 3)
circuit.cp(pi/3, 2, 3)
circuit.cp(pi/3, 2, 3)
circuit.cp(pi/3, 2, 3)
circuit.cp(pi/3, 2, 3)
circuit.barrier()
```

```python
circuit.swap(0, 2)
circuit.h(0)
circuit.cp(-pi/2, 0, 1)
circuit.h(1)
circuit.cp(-pi/4, 0, 2)
circuit.cp(-pi/2, 1, 2)
circuit.h(2)
circuit.barrier()
```

```python
circuit.measure(qr[0], cr[0])
circuit.measure(qr[1], cr[1])
circuit.measure(qr[2], cr[2])
```

```python
circuit.draw(output='mpl')
```

```python
#sim = Aer.get_backend('statevector_simulator')
sim = Aer.get_backend('qasm_simulator')
```

```python
res = execute(circuit, backend=sim).result()
```
