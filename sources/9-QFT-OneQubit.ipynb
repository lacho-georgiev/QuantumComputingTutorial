```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram, plot_bloch_multivector
```

```python
qc = QuantumCircuit(1)
```

```python
#qc.x(0)
qc.h(0)
```

```python
qc.draw(output='mpl')
```

```python
sim = Aer.get_backend('statevector_simulator')
```

```python
res = execute(qc, backend=sim).result()
```

```python
sv = res.get_statevector()
```

```python
plot_bloch_multivector(sv)
```
